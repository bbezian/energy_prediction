## Q1. Regular find function for list

```python
def find(values, target):
    if target in values:
        return values.index(target)
    else:
        return 'Element is not in the list'
```

#### Example to test the find function for Q1

```python
values = [6,1,4,9,7,10,15,9,17]
target= 9

print(f'Index of {target} is = {find(values, target)}')
```

## Q2. Optimised find function for sorted list

```python
def optimized_find(values, target): 
    
    left, right = 0, len(values)-1
    
    while right >= left:
        middle = left + (right - left)//2
        
        if values[middle] == target:
            return middle
        
        if values[middle] > target:
            right = middle -1
        else:
            left = middle + 1
    
    return 'Element is not in the list'
```

#### Example to test the optimized find function for Q2

```python
values = [1,4,6,7,9,10,15,16,17]
target= 7

print(f'Index of {target} is = {optimized_find(values, target)}')
```

```python
def statistical_indices(y_pred_trn,y_trn_1):
    RMSE_trn = sqrt(mean_squared_error(y_pred_trn, y_trn_1))
    R2_trn = (r2_score(y_pred_trn, y_trn_1)*100)
    Me_max_trn  = np.round(max(abs(y_pred_trn - y_trn_1)),2)
    MAPe_trn  = np.round(((sum(abs((y_pred_trn - y_trn_1)/y_trn_1)) * 100)/96),2)
    MBE_trn  = np.round((sum(abs(y_pred_trn - y_trn_1)))/(len(y_trn_1)),2)
    Tota_results_trn= pd.DataFrame((R2_trn, RMSE_trn, MAPe_trn, MBE_trn, Me_max_trn),
                                             columns=['ML'],index=['R2','RMSE','MAPe','MBE','Me_max'])
    return Tota_results_trn
```

## PCA Modeling

```python
n_components = 2

pca_train = PCA(n_components = n_components)

pca_data_train = pca_train.fit_transform(normalized_train)

df_pca_data_train= pd.DataFrame(data = pca_data_train, columns = [f'principal component {i+1}' for i in range(n_components)])

df_pca_data_train
```

```python
print("Clean Data")

print(f'{np.round(pca_train.explained_variance_ratio_,3)*100} --- % of variance explained by each component\
    \n{np.round(pca_train.singular_values_,3)} --- Singular values \
      \n\n Total variance explained by {n_components} components: {round(pca_train.explained_variance_ratio_.sum(),2)}')
```

## Find the matrix of covariance (Q)

```python
x = normalized_train.to_numpy()

pca = PCA()
pca.fit(x)

coeff = (np.transpose(pca.components_))

pd.DataFrame(coeff)

Q_train= coeff

pd.DataFrame(Q_train, index=['energy','enthalpy'], columns=['PC1','PC2'])
```

```python
PC_1 = []
PC_2 = []
for i in range(len(df_pca_data_train)):
    
    Q_1 = 0
    a = normalized_train['energy'].iloc[i]*Q_train[0,Q_1] + normalized_train['enthalpy'].iloc[i]*Q_train[1,Q_1]
    PC_1.append(a)
    
    Q_2 = 1
    b = normalized_train['energy'].iloc[i]*Q_train[0,Q_2] + normalized_train['enthalpy'].iloc[i]*Q_train[1,Q_2]
    PC_2.append(b)
    
```

```python
pd.DataFrame((PC_1,PC_2)).transpose()
```

## The score of i observation

```python
pd_F_score_train = pd.DataFrame((PC_1,PC_2), index = [f'score {i+1}' for i in range(2)]).transpose()

F_score_train = np.array(pd_F_score_train)

pd_F_score_train
```

## Standard deviation of scores (F) along the j-principal

```python
std_score_train = np.std(F_score_train, axis=0)

std_score_train
```

```python
n = 2 # The coefficient for the standard deviation 

threshold_train= []
for j in range (2):
    threshold_train.append(F_score_train[:,j]**2/(n*std_score_train[j])**2)
    
threshold_train = np.transpose(threshold_train)
pd_threshold_train= pd.DataFrame(threshold_train)
pd_threshold_train
```

## Creating the ellipsoid for abnormality detection

```python
t = np.arange(0,2*3.14,0.001)

x_ellipsoid_12_clean = []
y_ellipsoid_12_clean = []
for i in range (len(t)):
    x_ellipsoid_12_clean.append(math.cos(t[i]) * (n * std_score_train[0]))
    y_ellipsoid_12_clean.append(math.sin(t[i]) * (n * std_score_train[1]))

x_clean = np.transpose(x_ellipsoid_12_clean)
y_clean = np.transpose(y_ellipsoid_12_clean)

ellipsoid_clean = pd.DataFrame(x_clean, columns=['x_clean'])
ellipsoid_clean['y_raw']= y_clean
#ellipsoid_clean
```

## Bonus Question: Produce the merge_ranges function

```python
def merge_range(*List):
    combi = itertools.chain(*List)
    L = (set(combi))
    return [min(L) , max(L)]
```
